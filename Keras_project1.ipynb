```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

Example data 

1) An experimental drug was tested on individuals from age 13 to 100 in a 
clinical trial.
2) The trial had 2100 participants. Half were under 65 years old and half were above 65 years old.
3)Around 95% of the patients 65 years or older experienced side effects
4) Around 95% of the patients under 65 years experienced no side effects.


```python
for i in range(50):
  # for the ~5% of the younger patients who experienced side effetcs
  random_young = randint(13,64)
  train_samples.append(random_young)
  train_labels.append(1)

 # for the ~5% of the older patients who did not experience side effetcs
  random_old = randint(65,100)
  train_samples.append(random_old)
  train_labels.append(0)

 # for the ~95% of the younger patients who did not experience side effetcs
  random_young = randint(13,64)
  train_samples.append(random_young)
  train_labels.append(0) 

 # for the ~95% of the patients who experienced side effetcs
  random_old = randint(65,100)
  train_samples.append(random_old)
  train_labels.append(1)

  # 1 for patiemts who did experiece side effects and 0 for who did not.
```

```python
for i in train_samples:
  print(i)
```

```python
for i in train_labels:
  print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels,train_samples = shuffle(train_labels,train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
  print(i)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
#physical_devices = tf.config.experimental.list_physical_devices('GPU')
#print('Number of GPUs available: ', len(physical_devices))
#tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
model = Sequential([
    keras.layers.Dense(16, activation='relu', input_shape=(1,)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(2, activation='softmax')
    
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate =0.0001), loss='sparse_categorical_crossentropy', metrics =['accuracy'])

```

```python
model.fit(x=scaled_train_samples, y=train_labels, batch_size=10, epochs=40, shuffle=True, verbose=2)
```
